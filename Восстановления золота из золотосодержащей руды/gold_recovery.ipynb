Функция расчёта эффективности обогащения, по заданной в условии формуле:

```python
def get_recovery(c, f, t):
    result = np.nan
    
    try:
        result = 100 * c * (f - t) / (f * (c - t))
    except: 
        result = np.nan

    return result

def get_rougher_recovery(row):
    # C — доля золота в концентрате после флотации/очистки;
    c = row['rougher.output.concentrate_au']
    # F — доля золота в сырье/концентрате до флотации/очистки;
    f = row['rougher.input.feed_au']
    # T — доля золота в отвальных хвостах после флотации/очистки.
    t = row['rougher.output.tail_au']
    
    result = get_recovery(c, f, t)

    return result
```

```python
df_train['rougher.output.recovery_chk'] = df_train.apply(get_rougher_recovery, axis=1)
```

```python
df_train_rougher_recovery = df_train[['rougher.output.recovery', 'rougher.output.recovery_chk']]
```

```python
df_train_rougher_recovery.isnull().sum()
```

Удалим пропуски в данных, которые мешают расчёту ошибки

```python
df_train_rougher_recovery = df_train_rougher_recovery.dropna()
```

```python
train_rougher_recovery_mae = mean_absolute_error(
    df_train_rougher_recovery['rougher.output.recovery'], df_train_rougher_recovery['rougher.output.recovery_chk'])
```

## Модель

### Функция для вычисления итоговой sMAPE

```python
def get_smape(y_target, y_predicted):
    result = (2 * abs(y_target - y_predicted) / (abs(y_target) + abs(y_predicted))).mean()
    return result

def get_final_smape(rougher_target, rougher_predicted, final_target, final_predicted):
    result = 0.25 * get_smape(rougher_target, rougher_predicted) + 0.75 * get_smape(final_target, final_predicted)
    return result
```

### Функция оценки качества кроссвалидацией

```python
def get_crossvalidate_score(model, n_blocks, rougher_features, rougher_target, final_features, final_target):
    scores = []
    data_size = len(rougher_features)
    sample_size = int(data_size/n_blocks)
    indexes = rougher_features.index

    for i in range(0, data_size, sample_size):
        valid_indexes = list(indexes[i : i + sample_size])
        train_indexes = list(indexes[0 : i]) + list(indexes[i + sample_size:])

        rougher_features_train = rougher_features.loc[train_indexes]
        rougher_target_train = rougher_target.loc[train_indexes].iloc[:, 0].array
        
        rougher_features_valid = rougher_features.loc[valid_indexes]
        rougher_target_valid = rougher_target.loc[valid_indexes].iloc[:, 0].array

        final_features_train = final_features.loc[train_indexes]
        final_target_train = final_target.loc[train_indexes].iloc[:, 0].array
        
        final_features_valid = final_features.loc[valid_indexes]
        final_target_valid = final_target.loc[valid_indexes].iloc[:, 0].array

        model.fit(rougher_features_train, rougher_target_train)
        rougher_predicted_valid = model.predict(rougher_features_valid)
        
        model.fit(final_features_train, final_target_train)
        final_predicted_valid = model.predict(final_features_valid)
     
        score = get_final_smape(rougher_target_valid, rougher_predicted_valid, final_target_valid, final_predicted_valid)

        scores.append(score)
 
    return pd.Series(scores).mean()
```
